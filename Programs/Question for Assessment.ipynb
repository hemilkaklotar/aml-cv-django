Write a program to generate all possible combinations of the word "History". For example, if I give input as CAT then the answer would be CAT, CTA, ACT, ATC, TAC, TCA. Note: Any language is preferable. Please upload your code on Dropbox or Google Drive and share the link in the answer.

```python
def combination(s, answer):
    if (len(s) == 0):
        print(answer, end = "  ")
        return
     
    for i in range(len(s)):
        ch = s[i]
        left_substr = s[0:i]
        right_substr = s[i + 1:]
        rest = left_substr + right_substr
        combination(rest, answer + ch)
 

answer = ""
 #input is History
s = input("Enter the string : ")
 
print("All possible strings are : ")
combination(s, answer)
```
